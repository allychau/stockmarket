```python
import pandas as pd
import numpy as np
```

```python
#Load Raw Data
df = pd.read_csv("stocks.csv")
df
```

```python
# Task 1: Split stock per stock ticker
df1 = df[df['ticker']=="GOOGL"]
df1
```

```python
# Sub-Task 1: Introducing calculated column volatility
df1['volatility-calc'] = df1.high - df1.low
```

```python
# Sub-Task 2: Introducing calculated column 20-Day and 100-day SMA
df1['SMA-20-calc']=df.iloc[:,13].rolling(window=20).mean()
df1['SMA-100-calc']=df.iloc[:,13].rolling(window=100).mean()
```

```python
# Sub-Task 3: Introducing Rate of Change (Price, Volume, Volatility)
df1['ROC_Price']=df1['close'].diff()
df1['ROC_Volume']=df1['adj_volume'].diff()
df1['ROC-Volatility']=df1['volatility-calc'].diff()
df1
```

```python
# Sub-Task 4: Add crossover column to check 20-day vs. 100-day SMA

df1['Crossover']=""

for index, row in df1.iterrows():
    if (row['SMA-20-calc'] > row['SMA-100-calc']):
        df1.loc[index, 'Crossover']=1
    else:
        df1.loc[index, 'Crossover']=0
df1
```
