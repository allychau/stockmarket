```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
#Load Raw Data
df = pd.read_csv("stocks.csv")
#df
```

```python
# Task 1: Split stock per stock ticker
df1 = df[df['ticker']=="GOOGL"]
df1.drop(df.columns[[15, 16, 17, 18]], axis=1, inplace=True)
```

```python
# Task 2: Introducing calculated column volatility
df1['volatility-calc'] = df1.high - df1.low
```

```python
# Task 3: Introducing calculated column 20-Day and 100-day SMA
df1['SMA-10-calc']=df.iloc[:,13].rolling(window=10).mean()
df1['SMA-50-calc']=df.iloc[:,13].rolling(window=50).mean()
```

```python
# Task 4: Introducing Rate of Change (Price, Volume, Volatility)
df1['ROC_Price']=df1['close'].diff()
df1['ROC_Volume']=df1['adj_volume'].diff()
df1['ROC_Volatility']=df1['volatility-calc'].diff()
```

```python
# Task 5: Add crossover column to check 10-day vs. 50-day SMA

df1['Crossover']=""

for index, row in df1.iterrows():
    if (row['SMA-10-calc'] > row['SMA-50-calc']):
        df1.loc[index, 'Crossover']=1
    else:
        df1.loc[index, 'Crossover']=0      
```

```python
# Task 6A (Optional): Reformat date cells to yyyy-mm-dd
df1['date']=pd.to_datetime(df['date'])  
```

```python
# Task 6B: Fill NaN values with column means so ML model can run with usable data

df1['SMA-10-calc'].fillna(value=df1['SMA-10-calc'].median(), inplace=True)
df1['SMA-50-calc'].fillna(value=df1['SMA-50-calc'].median(), inplace=True)
df1['ROC_Price'] = df1['ROC_Price'].fillna(0)
df1['ROC_Volume'] = df1['ROC_Price'].fillna(0)
df1['ROC_Volatility'] = df1['ROC_Price'].fillna(0)
df1.head(60)
```

```python
# Task 7: Begin to run ML

# Define x,y to get ready for ML linear regression
X = df1[["adj_volume","volatility-calc","Crossover"]]
y = df1["adj_close"].values.reshape(-1, 1)
print(X.shape, y.shape)
#putting data in a way the ML algorithm can use
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
X_train.head()

#defining the training/testing data and algortihm randomizes what is the training/testing data. you do not
#have to specify it. 
```

```python
from sklearn.preprocessing import StandardScaler
X_scaler = StandardScaler().fit(X_train)
y_scaler = StandardScaler().fit(y_train)

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
y_train_scaled = y_scaler.transform(y_train)
y_test_scaled = y_scaler.transform(y_test)

#X = df1[["20-Day SMA", "100-Day SMA", "Volatility", "Lag Gap"]].values.reshape(-1,1)
#y = df1["close"].values.reshape(-1, 1)
#print(X.shape, y.shape)

#scaling the data and re-shape just to be sure
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train_scaled, y_train_scaled)
plt.scatter(model.predict(X_train_scaled), model.predict(X_train_scaled) - y_train_scaled, c="blue", label="Training Data")
plt.scatter(model.predict(X_test_scaled), model.predict(X_test_scaled) - y_test_scaled, c="orange", label="Testing Data")
plt.legend()
plt.hlines(y=0, xmin=y_test_scaled.min(), xmax=y_test_scaled.max())
plt.title("Residual Plot")
plt.show()

# running the linear regression model and plotting residuals
```

```python
from sklearn.metrics import mean_squared_error

predictions = model.predict(X_test_scaled)
MSE = mean_squared_error(y_test_scaled, predictions)
r2 = model.score(X_test_scaled, y_test_scaled)

print(f"MSE: {MSE}, R2: {r2}")

#fit based on all 4 variables in X axis
```
